```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
```

```python
plt.bar(carprice.CarName,carprice.price)
plt.xlabel('CarName')
plt.ylabel('price')
plt.title('Bar Plot')
plt.show()
```

```python
plt.boxplot(carprice.price)
plt.xlabel('Price')
plt.ylabel('Number of cars')
plt.title('Box Plot')
plt.show()
```
